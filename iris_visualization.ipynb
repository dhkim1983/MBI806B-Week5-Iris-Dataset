```python
# Week 5 Activity 1 Part 2
# Iris Dataset Visualization

!pip -q install ucimlrepo

from ucimlrepo import fetch_ucirepo
import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset from UCI Repository
iris = fetch_ucirepo(id=53)

# Get the target/class column
y = iris.data.targets

# Count the number of samples for each flower class
class_counts = y['class'].value_counts()

print("Number of samples for each Iris class:")
print(class_counts)

# Create a bar chart
plt.figure(figsize=(8, 5))
class_counts.plot(kind='bar')

plt.title('Number of Samples for Each Iris Flower Class')
plt.xlabel('Iris Flower Class')
plt.ylabel('Number of Samples')
plt.xticks(rotation=20)
plt.tight_layout()

# Save the result image
plt.savefig('iris_class_distribution.png')

plt.show()

# Check whether the classes are balanced
if class_counts.nunique() == 1:
    print("The classes are balanced because each class has the same number of samples.")
else:
    print("The classes are not balanced because the sample counts are different.")
```
